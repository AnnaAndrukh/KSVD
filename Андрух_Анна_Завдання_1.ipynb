```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, Embedding, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
```

```python
# Text data for training
texts = ["Текст для навчання", "Дуже сподобався цей текст", "Можна писати ще", "Пишуть ще", "Цей сайт не працює",
         "Треба налаштувати сайт", "Вчора було весело", "Хочу це робити, бо цікаво"]

# Preprocessing text
texts = [text.replace("увімкнено", '') for text in texts]  # Clean the text if needed

# Tokenize the texts
maxWordsCount = 1000
tokenizer = Tokenizer(num_words=maxWordsCount, filters="!\"#$%&'()*+,-./:;<=>?@[\\]^_`{|}~\t\n", lower=True, split=' ', char_level=False)
tokenizer.fit_on_texts(texts)

# Convert text to sequences
data = tokenizer.texts_to_sequences(texts)

inp_words = 3  # Number of words in each sequence
X = []
Y = []

# Create sequences for training
for i in range(len(data)):
    for j in range(len(data[i]) - inp_words):
        X.append(data[i][j:j + inp_words])
        Y.append(data[i][j + inp_words])

X = np.array(X)
Y = np.array(Y)

# Convert labels to categorical
Y = to_categorical(Y, num_classes=maxWordsCount)

# Build the RNN model
model = Sequential()
model.add(Embedding(maxWordsCount, 300, input_length=inp_words))  # Embedding layer
model.add(LSTM(128, activation='tanh', return_sequences=True))  # LSTM layer
model.add(LSTM(128, activation='tanh'))  # Another LSTM layer
model.add(Dense(maxWordsCount, activation='softmax'))  # Output layer

# Compile the model
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')

# Train the model
history = model.fit(X, Y, batch_size=32, epochs=100)

# Plotting the loss and accuracy graphs
plt.figure(figsize=(12, 4))

# Plotting loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Function')
plt.legend()

# Plotting accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Accuracy', color='green')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy')
plt.legend()

plt.show()

# Model summary
model.summary()

# Function to generate new text
def buildPhrase(texts, str_len=20):
    data = tokenizer.texts_to_sequences([texts])[0]
    for i in range(str_len):
        x = np.array(data[-inp_words:])
        x = np.expand_dims(x, axis=0)
        pred = model.predict(x)
        index = np.argmax(pred)
        data.append(index)
    res = ' '.join([tokenizer.index_word.get(i, '') for i in data])
    return res

# Generating new text
res = buildPhrase("Хочу", str_len=20)
print(res)
```
